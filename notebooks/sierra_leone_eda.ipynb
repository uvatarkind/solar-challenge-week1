```python
# SIERRALEONE EDA NOTEBOOK

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore
```

```python
# Load dataset
sierraleone_df = pd.read_csv("data/sierraleone-bumbuna.csv", parse_dates=["Timestamp"])
```

```python
# Basic info
display(sierraleone_df.head())
display(sierraleone_df.info())
display(sierraleone_df.describe())
```

```python
# Missing values
missing = sierraleone_df.isna().mean() * 100
print("Missing over 5%:\n", missing[missing > 5])

```

```python
# Outlier detection
num_cols = ["GHI", "DNI", "DHI", "TModA", "TModB", "WS", "WSgust"]
z_scores = sierraleone_df[num_cols].apply(zscore)
outliers = (z_scores.abs() > 3).sum()
print("Outliers:\n", outliers)
```

```python
# Cleaning: Fill missing values
sierraleone_df[num_cols] = sierraleone_df[num_cols].fillna(sierraleone_df[num_cols].median())
```

```python
# Drop extreme outliers
sierraleone_df = sierraleone_df[(z_scores.abs() <= 3).all(axis=1)]
```

```python
# Save cleaned version
sierraleone_df.to_csv("data/sierraleone_clean.csv", index=False)

# Add time features
sierraleone_df["month"] = sierraleone_df["Timestamp"].dt.month

# Time series plot
sierraleone_df.set_index("Timestamp")[["GHI", "DNI", "DHI"]].plot(figsize=(15, 6))
plt.title("Solar Data Over Time - Sierraleone")
plt.grid()
```

```python

# Monthly averages
sierraleone_df.groupby("month")[["GHI", "DNI", "DHI", "Tamb"]].mean().plot(kind="bar")

# Cleaning impact
sierraleone_df.groupby("Cleaning")[["ModA", "ModB"]].mean().plot(kind="bar")

# Correlation
sns.heatmap(sierraleone_df[num_cols].corr(), annot=True, cmap="coolwarm")

# Scatter plots
sns.scatterplot(data=sierraleone_df, x="RH", y="Tamb")
sns.scatterplot(data=sierraleone_df, x="WS", y="GHI")

# Bubble chart
plt.scatter(sierraleone_df["GHI"], sierraleone_df["Tamb"], s=sierraleone_df["RH"], alpha=0.5)
plt.xlabel("GHI")
plt.ylabel("Tamb")
plt.title("GHI vs Tamb (bubble size = RH)")
```
